```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x= [2,4,6]
y= [2,4,7]
```

```python
x_array= np.array(x)
y_array= np.array(y)
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
print(x_mean, y_mean)
```

```python
x_correlation = np.corrcoef(x_array,y_array)
print(x_correlation)
```

```python
plt.title("Correlation")
plt.scatter(x_array, y_array)
plt.plot(np.unique(x_array), np.poly1d(np.polyfit(x_array, y_array, 1))(np.unique(x_array)), color = "red")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```
